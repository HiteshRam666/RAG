### **Vector Database**

```python
from langchain_community.document_loaders import TextLoader 
from langchain_openai import OpenAIEmbeddings 
from langchain_text_splitters import CharacterTextSplitter 
from langchain_chroma import Chroma 
```

```python
loaded_document = TextLoader("data/state_of_the_union.txt", encoding='utf-8').load()
```

```python
text_splitter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 0)
```

```python
chunks_of_text = text_splitter.split_documents(loaded_document)
```

```python
len(chunks_of_text)
```

```python
vector_db = Chroma.from_documents(chunks_of_text, OpenAIEmbeddings())
```

```python
vector_db
```

```python
question = "What did president say about the John Lewis voting rights act ?"

response = vector_db.similarity_search(question) 
```

```python
print(response[0].page_content)
```

#### **Vector Store as Retriever** 
- Find the embedding that best answers your question

```python
from langchain_community.document_loaders import TextLoader 

loader = TextLoader("data/state_of_the_union.txt", encoding='utf-8')
```

```python
from langchain_community.vectorstores import FAISS 
from langchain_openai import OpenAIEmbeddings 
from langchain_text_splitters import CharacterTextSplitter 

loaded_document = loader.load() 
```

```python
text_splitter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 0)
```

```python
chunk_of_text = text_splitter.split_documents(loaded_document)
```

```python
embeddings = OpenAIEmbeddings()
```

```python
vector_db = FAISS.from_documents(chunk_of_text, embeddings)
```

```python
vector_db
```

```python
# Creating Retriever
retriever = vector_db.as_retriever(search_kwargs = {'k': 1}) 
```

```python
response = retriever.invoke("What did he say about ketanji brown jackson ?")
response
```

```python
response[0].page_content
```
